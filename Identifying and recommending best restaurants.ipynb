## Loading the excel file i.e data

```python
data = pd.read_excel("data.xlsx")
data.head()
```

## Loading another excel file i.e Country-code

```python
cc=pd.read_excel("Country-Code.xlsx")
cc.head()
```

## Merging the two dataframes i.e (data and cc) based on a common column which is Country code using LEFT JOIN (how="left")

```python
df_rest = pd.merge(data,cc,on='Country Code', how ='left')
df_rest.head()
```

## Adding ('_') for all the column names using Replace() method

```python
df_rest.columns = df_rest.columns.str.replace(" ", "_")
df_rest.columns
```

```python
df_rest.info()
```

# EDA - 2

Explore how ratings are distributed overall.

Evaluate the Highest Rated and Lowest Rated Restaurant of the City in all the countries.

Restaurant franchise is a thriving venture. So, it becomes very important to explore the franchise with most national presence.

```python
max_rate = df_rest.groupby(["Country","City"]).agg(max_rating = ("Aggregate_rating", "max"))
max_rate.sort_values(by='max_rating', ascending=False)
```

```python
min_rate = df_rest.groupby(["Country","City"]).agg(min_rating = ("Aggregate_rating", "min"))
min_rate.sort_values(by='min_rating', ascending=False)
```

```python
# highest_rating restaurants
max_rate = df_rest.sort_values(by="Aggregate_rating",ascending=False).groupby(["Country","City"],as_index=False).first()

# lowest rating restaurants 
min_rate = df_rest.sort_values(by="Aggregate_rating",ascending=False).groupby(["Country","City"],as_index=False).last()

#new dataframe created for high rated restaurants
df_max = max_rate[["Country","City","Restaurant_Name","Aggregate_rating"]] 

#new dataframe created for low rated restaurants
df_min = min_rate[["Country","City","Restaurant_Name","Aggregate_rating"]]

# merge both dataframes into single dataframe
rating_rest=df_max.merge(df_min,left_on="City",right_on="City",how="inner")

rating_rest
```

### Dropping Country_y 

```python
rating_rest.drop(columns="Country_y",axis=1,inplace=True)
rating_rest.columns=["Country","City", "Highest Rated Restaurant", "Rating Max","Lowest Rated Restaurant","Rating Min"]
rating_rest
```

#### We observe that India and USA has the highest no of restaurants , we will try to see the distribution of restaurants ratings for these two countries.

# EDA - 3
Ratio between restaurants that allow table booking vs that do not allow table booking.

Percentage of restaurants providing online delivery.

Difference in no. of votes for the restaurants that deliver and the restaurant that dont.

```python
#### Ratio between restaurants that allow table booking vs that do not allow table booking.


table_booking = df_rest[df_rest['Has_Table_booking']=="Yes"]['Restaurant_ID'].count()
table_nbooking = df_rest[df_rest['Has_Table_booking']=="No"]['Restaurant_ID'].count()
print('Ratio between restaurants that allow table booking vs that do not allow table booking is: ',  round((table_booking/table_nbooking),2))
```

```python
print(table_booking,table_nbooking)
```

```python
# Pie Chart to show percentage of restaurants providing table booking vs no table booking 

labels = 'Table Booking', 'No Table Booking'
palette_color = sns.color_palette('bright') 
explode = (0.1,0)

plt.title("Table Booking vs No Table Booking")
plt.pie(df_rest["Has_Table_booking"].value_counts(), explode=explode, labels = labels,autopct='%1.1f%%') 
plt.show() 
```

```python
len(df_rest)
```

```python
##Percentage of restaurant that provides online delivery

online_delivery = df_rest[df_rest['Has_Online_delivery']=="Yes"]['Restaurant_ID'].count()
online_ndelivery = df_rest[df_rest['Has_Online_delivery']=="No"]['Restaurant_ID'].count()
print('Percentage of restaurants providing online delivery : {} %'.format((round(online_delivery/len(df_rest),3)*100)))
```

```python
# Pie Chart to show percentage of restaurants providing online delivery vs no online delivery 

labels = 'Online Delivery','No Online Delivery'
palette_color = sns.color_palette('bright') 
explode = (0.1,0)

plt.title("Online Delivery vs No Online Delivery")
plt.pie(df_rest["Has_Online_delivery"].value_counts(),explode=explode,labels = labels,autopct='%1.1f%%') 
plt.show() 
```

```python
# Difference in no. of votes for the restaurants that deliver and the restaurant that dont.

online_deliver = df_rest[df_rest['Has_Online_delivery'] == "Yes"]['Votes'].sum()
online_ndeliver = df_rest[df_rest['Has_Online_delivery'] == "No"]['Votes'].sum()
print('Difference in no. of votes for the restaurants that deliver and the restaurant that dont: ',abs((online_deliver - online_ndeliver)))
```

#### We observe that, out of the total votes about 25.7% votes were given to restaurants that don't have online delivery option

#### Out of the total votes about 74.3% votes were given to restaurants that have online delivery option

#### This clearly shows that restaurants that have online delivery are more likely to get a vote(feedback)

**Explain the factors in the data that may have an effect on ratings. For example, number of cuisines, cost, delivery option, etc.?**

```python
plt.figure(figsize = (7,3))

sns.regplot(y='Average_Cost_for_two',x='Aggregate_rating',data=df_rest)
df_rest[["Average_Cost_for_two", "Aggregate_rating"]].corr()
```

We can see Average rating and Aggregated rating has no correlation 

```python
plt.figure(figsize = (7,3))

sns.regplot(y='Votes',x='Aggregate_rating',data=df_rest)
df_rest[["Votes", "Aggregate_rating"]].corr()
```

We can see Votes and Aggregated rating has a positive correlation

```python
plt.figure(figsize = (7,3))

sns.regplot(y='Price_range',x='Aggregate_rating',data=df_rest)
df_rest[["Price_range", "Aggregate_rating"]].corr()
```

We can see Price range and Aggregated rating has a positive correlation

```python
plt.figure(figsize = (7,3))

sns.regplot(x='Has_Table_booking',y='Aggregate_rating',data=df_rest)
df_rest[["Has_Table_booking", "Aggregate_rating"]].corr()
```

We see that there is no single variable that affects the rating strongly, however table booking,online delivery,avg price for two and price range, number of votes do play a part in affecting the rating of a restaurant.
